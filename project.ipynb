#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph? You must use the subset obtained in the question 14 (★★☆) 
Hint: make a scatter plot of each population using subplots.

```python
import matplotlib.pyplot as plt

normalized_prices = []
populations_list = []
populations = subset['level5'].unique()

# Iterate over each unique population
for cur_population in populations:
    # Create a subset for the current population
    cur_subset = subset[subset['level5'] == cur_population][['price']]
    
    # Normalize the 'price' column within the subset using max normalization
    max_abs = cur_subset['price'].abs().max()
    if max_abs != 0:
        cur_normalized_prices = cur_subset['price'] / max_abs
        
        # Store normalized prices and corresponding population labels
        normalized_prices.extend(cur_normalized_prices)
        populations_list.extend([cur_population] * len(cur_normalized_prices))
        print(normalized_prices)
        print(populations_list)

# Create scatter plot for normalized prices
plt.figure(figsize=(10, 6))
plt.scatter(populations_list, normalized_prices, alpha=0.5)

# Add labels and title
plt.xlabel('Population')
plt.ylabel('Normalized Price')
plt.title('Scatter Plot of Normalized Prices Across All Populations')

plt.grid(True)
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in the question 14) (★★★★)

Execute the following cell and then start coding in the next one. You must implement a simple code that transform the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
import folium
m = folium.Map(location = [40.317715, -3.784103393554688], zoom_start= 10)
m
```

```python
## HERE: plot the coordinates of the estates

population = subset['level5'].unique()
population_colors = {'Fuenlabrada': 'red', 'Alcorcón': 'blue', 'Leganés': 'green', 'Getafe': 'orange'}
for cur_population in population:
        color = population_colors.get(cur_population, 'black')
        cur_subset = subset[subset['level5'] == cur_population][['longitude','latitude','level5']]
        cur_subset['color'] = color
        cur_subset['latitude'] = cur_subset['latitude'].str.replace(',', '.')
        cur_subset['longitude'] = cur_subset['longitude'].str.replace(',', '.')
        cur_subset['latitude'] = cur_subset['latitude'].astype(float)
        cur_subset['longitude'] = cur_subset['longitude'].astype(float)
        for i in range(len(cur_subset)):
                        folium.Marker(
                        location=[cur_subset.iloc[i]['latitude'], cur_subset.iloc[i]['longitude']], 
                        popup=cur_subset.iloc[i]['level5'],
                        icon=folium.Icon(color=color)
                        ).add_to(m)
                
```
